```python
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd
```

```python
url="https://keithgalli.github.io/web-scraping/webpage.html"
headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
web_page=requests.get(url,headers=headers,timeout=3)
web_page.raise_for_status()
```

```python
soup=BeautifulSoup(web_page.content,"lxml")
print(soup)
```

```python
web_table=soup.find("table",attrs={"class":"hockey-stats"})
print(web_table)
```

```python
web_table=soup.select("table.hockey-stats")[0]
print(web_table)
```

```python
web_table_columns=soup.find("thead").find_all("th")
print(web_table_columns)
```

```python
web_table_column_names=[c.string for c in web_table_columns]
print(web_table_column_names)
```

```python
web_table_rows=soup.find("tbody").find_all("tr")
web_data=[]
for table_rows in web_table_rows:
  table_data=table_rows.find_all("td")
  row=[str(table_rows.get_text()).strip() for table_rows in table_data]
  web_data.append(row)
  print(web_data[0])
```

```python
df=pd.DataFrame(web_data,columns=web_table_column_names)
df.head()
```
